Thanks to Power Bi, we can easily display the trend for the snow precipitation and the ground accumulation.

Looking at the past 100 years, the amount of snowfall has reduced by about <b>27%</b> from its peak of 296 cm in 1927. 
<br>The amount does seem to have stabilized since the early 90s with a dip occuring only in 2007.

The data for snow accumulation was limited so we can only go back 76 years. <br>
It looks like the amount has reduced by about <b>67%</b> from its peak of 10.15 cm in 1956 and the trend could continue. 

We can now conclude that the amount of snow has reduced progressively as the years go on. The accumulated snow specifically has drastically reduced probably due to a combination of climate change and snow removal proficiency. <br>Further sources of data would be required for more elaborate answers.
